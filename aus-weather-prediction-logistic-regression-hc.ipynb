```python
# Intialise the values of w and b 

def initialise(n_features):
    w = np.random.randn(n_features)*0.01
    return w , 0
```

```python
# I Basically Tend to use Normal Distribution method 

def normalise_data(data, mean=None, std=None):
    if mean is None:
        mean = np.mean(data, axis=0)
    if std is None:
        std = np.std(data, axis=0)

    std[std == 0] = 1  
    data = (data - mean) / std  

    return data, mean, std
 

# For even having the boundary nit always to be a straight line to fint the boundary ti the data more perfectly 

def poly(data,degree):
    poly = PolynomialFeatures(degree= degree ,include_bias= False)
    return poly.fit_transform(data)
```

```python
# calculate wx+b value 

def compute_f(data, w ,b):
    return np.dot(data,w) +b


# calculate Sigmoid Function_value 

def sigmoid(f_data):
    sigmoid_data = 1 / (1 + np.exp(-f_data))
    return sigmoid_data

```

The loss function above can be rewritten to be easier to implement.
    $$loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = (-y^{(i)} \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)$$
  
This is a rather formidable-looking equation. It is less daunting when you consider $y^{(i)}$ can have only two values, 0 and 1. One can then consider the equation in two pieces:  
when $ y^{(i)} = 0$, the left-hand term is eliminated:
$$
\begin{align}
loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), 0) &= (-(0) \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - 0\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) \\
&= -\log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)
\end{align}
$$
and when $ y^{(i)} = 1$, the right-hand term is eliminated:
$$
\begin{align}
  loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), 1) &=  (-(1) \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - 1\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)\\
  &=  -\log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right)
\end{align}
$$

OK, with this new logistic loss function, a cost function can be produced that incorporates the loss from all the examples. This will be the topic of the next lab. For now, let's take a look at the cost vs parameters curve for the simple example we considered above:

```python
# We compute loss function isntead of cost function 

def compute_cost_function(sigmoid_data,target_data,w,lambda_):
    m = sigmoid_data.shape[0]
    return (1/m)* np.sum((-target_data*np.log(sigmoid_data)) - (1- target_data)*np.log(1-sigmoid_data)) + (lambda_/2*m)*np.sum(w**2)


def compute_gradient(sigmoid_data,data,target_data,lambda_):
    m = sigmoid_data.shape[0]
    error = sigmoid_data-target_data
    dj_dw = np.dot(error,data) * (1/m)
    dj_db = np.sum(error)/m
    return dj_dw , dj_db

```

```python
def predict(data , w , b):
    z = np.dot(data,w)+b
    return sigmoid(z)
```

```python
def gradient(data ,target ,w_initial ,b_initial ,iterations ,alpha , lambda_):
    j_history = []

    w = w_initial

    b = b_initial

    diff = float('inf')

    i = 0

    cost = compute_cost_function(sigmoid(compute_f(data,w,b)),target,w,lambda_)

    j_history.append(cost)

    sigmoid_data = sigmoid(compute_f(data, w, b))

    while (i < iterations and diff > 1e-15):

        sigmoid_data = sigmoid(compute_f(data,w,b))

        dj_dw , dj_db = compute_gradient(sigmoid_data,data,target,lambda_)

        w -= alpha* dj_dw

        b -= alpha* dj_db

        cost = compute_cost_function(sigmoid(compute_f(data,w,b)),target,w,lambda_)

        j_history.append(cost)

        i+=1

        diff = abs(j_history[-1] - j_history[-2])

    return sigmoid_data ,w , b , j_history 

```

```python
m , n = test_data.shape

w , b = initialise(n)

print(f" Initilaised the values of  ")

store_model_results = []

best_model = None

alpha_values= [0.1,0.01,0.001]

iterations_values = [500 ,1000, 10000]

train_data , mean , std = normalise_data(train_data)

test_data , mean , std = normalise_data(test_data)

lambda_values = [0.1,0.01,0.001]

for degree in range(1, 4):

    for alpha in alpha_values :

        for iterations in iterations_values:

            for lambda_ in lambda_values:

                print(f"Running Testcases on degree : {degree} , alpha : {alpha} , iterations = {iterations} , lambda : {lambda_}")

                record ={'degree' : degree , 'alpha' : alpha , 'iterations' : iterations}
                
                record['sigmoid_data'], record['w'], record['b'] , record['j_history'] = gradient(train_data,train_data_labels,w,b,iterations,alpha,lambda_)

                test_predictions = predict(test_data, record['w'], record['b'])

                test_predictions = (test_predictions >= 0.5).astype(int)

                accuracy = accuracy_score(test_data_labels, test_predictions)

                record['accuracy'] = accuracy 

                accuracy = accuracy * 100

                print(f"Model Accuracy: {accuracy:.5f}% ")

                store_model_results.append(record)

                if best_model is None :
                    
                    best_model = record 
                
                else :

                    if record['accuracy'] > best_model['accuracy']:

                        best_model = record
                

                
                

            


```
